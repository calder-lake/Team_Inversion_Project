```python
from openpyxl import load_workbook
import csv
import os
import ctypes
from openpyxl import load_workbook
import pandas as pd

input_file = r"\\pesamba.int.chpc.utah.edu\PE-projects3-1\proj_tnx_ops\Proj_Caitlin Roake\patient.csv"

df_sample = pd.read_csv(
    input_file,
    skiprows=lambda i: i > 0 and __import__("random").random() > 0.001
)

print(df_sample.head())
print(len(df_sample))

patient_ids = set(df_sample["patient_id"])



input_file = r"\\pesamba.int.chpc.utah.edu\PE-projects3-1\proj_tnx_ops\Proj_Caitlin Roake\diagnosis.csv"
output_file = r"\\pesamba.int.chpc.utah.edu\PE-projects3-1\proj_tnx_ops\Proj_Caitlin Roake\diagnosis_filtered.csv"

chunksize = 100_000  # adjust if needed
first_chunk = True

for chunk in pd.read_csv(input_file, chunksize=chunksize, dtype=str):
    filtered = chunk[chunk["patient_id"].isin(patient_ids)]
    
    if not filtered.empty:
        filtered.to_csv(
            output_file,
            mode="w" if first_chunk else "a",
            header=first_chunk,
            index=False
        )
        first_chunk = False

print("Done.")


```
